```python
num_cols1=data1[['weight',
                 'gender',
                 'ball_controlling_skills',
                 'jumping_skills','strong_foot',
                 'behaviour_rating',
                 'matches_played',
                 'fitness_rating',
                 'coaching',
                 'years_of_experience',
                 'jumping_skills',
                  'penalties_conversion_rate',
                  'mental_strength']]
```

```python
num_cols1.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
## using stratergy most frequent in simple imputer function handling  missing null values
imputer =SimpleImputer(missing_values =np.nan,strategy='most_frequent')
```

```python
imputer = imputer.fit(num_cols1) ##fit simple imputer functio
```

```python
num_cols1 = imputer.transform(num_cols1) ## transform all data (null) in the
```

```python
type(num_cols1)
```

```python
num_cols1= pd.DataFrame(num_cols1,columns=['weight',
                 'gender',
                 'ball_controlling_skills',
                 'jumping_skills','strong_foot',
                 'behaviour_rating',
                 'matches_played',
                 'fitness_rating',
                 'coaching',
                 'years_of_experience',
                 'jumping_skills',
                  'penalties_conversion_rate',
                  'mental_strength'])
```
